```python
tStart = time.time()
path = 'C:\\Users\\ecom\\Desktop\\NBAData'
if not os.path.isdir(path):
    os.mkdir(path)

for i in year:
    path = 'C:\\Users\\ecom\\Desktop\\NBAData\\' + str(i)
    if not os.path.isdir(path):
        os.mkdir(path)
        
    for t in range(len(nbateam)):
        if t == 0:
            s = 'East'
        elif t == 1:
            s = 'West'
        path = 'C:\\Users\\ecom\\Desktop\\NBAData\\' + str(i) + '\\' + s
        if not os.path.isdir(path):
            os.mkdir(path)
        path1 = path
        for j in nbateam[t]:
            
            #change teamname
            teamname = j
            if i <= 2012 and j == 'BRK':
                j = 'NJN'
            elif i <= 2014 and j == 'CHO':
                j = 'CHA'
            elif i == 2007 and j == 'NOP':
                j = 'NOK'
            elif i >= 2008 and i <= 2013 and j == 'NOP':
                j = 'NOH'
            elif i <= 2008 and j == 'OKC':
                j = 'SEA'
                
            
            path = path1 + '\\' + teamname
            if not os.path.isdir(path):
                os.mkdir(path)
            webplayerlist = []
            url = 'https://www.basketball-reference.com/teams/' + j + '/' + str(i) + '.html'
#             print(url)
            table = pd.read_html(url)
            playerlist = list(table[0].Player)
            for k in range(len(playerlist)):
                count = 0
                st = ''
                st1 = ''
                for l in playerlist[k]:
                    if l == '.':
                        continue
                    else:
                        count = count + 1
                        st = st + l
                        if count <= 2:
                            st1 = st1 + l
                        elif l == ' ':
                            st = ''
                st2 = ''
                for w in st:
                    if len(st2) < 5 and w != '\'':
                        st2 = st2 + w
                st = st2 + st1
                st = st.lower()
                webplayerlist.append(st)
            Players = pd.DataFrame(columns=['name', 'webname'])
            Players.name = playerlist
            Players.webname = webplayerlist
            Players.to_excel(path1 + '\\' + teamname + '\\' + teamname + '.xlsx')
tEnd = time.time()
print('Cost %f s' %(tEnd - tStart))
#almost 20 mins
```
