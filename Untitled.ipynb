```python
import os, ast

def menu():
    print("id, password manage system")
    print("-----------------------------------")
    print("1.enter id and password")
    print("2.show id and password")
    print("3.change password")
    print("4.delete id and password")
    print("0.end program")
    print("-----------------------------------")
def ReadData():
    with open('password.txt', 'r', encoding = 'UTF-8-sig') as f:
        filedata = f.read()
        if filedata != "":
            data = ast.literal_eval(filedata)
            return data
        else:
            return dict()
def disp_data():
    print("id\t password")
    print("==================")
    for key in data:
        print("{}\t{}".format(key,data[key]))
    input("press any key to return main selection")
def input_data():
    while True:
        name = input("please enter id (Enter ==> stop entering) : ")
        if name == "":
            break
        if name in data:
            print("{} this id is exists".format(name))
            continue
        password = input("please enter password : ")
        data[name] = password
        with open('password.txt', 'w', encoding = 'UTF-8-sig') as f:
            f.write(str(data))
        print("{} is already stored".format(name))
def edit_data():
    while True:
        name = input("please enter the id that you want to change(Enter) ===> stop entering : ")
        if name == "":
            break
        if not name in data:
            print("{} id is not exists!".format(name))
            continue
        print("the original password is {}".format(data[name]))
        password = input("please enter new password : ")
        data[name] = password
        with open('password.txt', 'w', encoding = "UTF-8-sig") as f:
            f.write(str(data))
            intput("password is changed. press any key to go back main menu")
            break
def delete_data():
    while True:
        name = input("please enter the id that you want to delete(Enter ==> stop entering) : ")
        if not name in data:
            print("{} is not exists!".format(name))
            continue
        print("delete {}\'s data".format(name))
        yn = intput("(Y/N)?")
        if(yn == "Y" or yn == "y"):
            del data[name]
            with open('password.txt', 'w', encoding = 'UTF-8-sig') as f:
                f.write(str(data))
                input("delete conplete, please press any key to go back main menu")
                break
data = dict()

data = ReadData()
while True:
    menu()
    choice = int(input("please enter your choice : "))
    print()
    if choice == 1:
        input_data()
    elif choice == 2:
        disp_data()
    elif choice == 3:
        edit_data()
    elif choice == 4:
        delete_data()
    else:
        break
print("progeam is run to the end!")
```
