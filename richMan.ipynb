# 取得中古屋網址

```python
num = []
webfront = 'https://sale.591.com.tw/?shType=list&role=1&regionid=1&firstRow='
webafter = '&totalRows=700'
urlQueue = queue.Queue()
for number in range(0, 700, 30):
    num.append(number)
for i in num:
    url = webfront + str(i) + webafter
    urlQueue.put(url)
    print(url)
```

```python
link = []
link3 = []
def getwebsite(n):
    while True:
        try:
            url = n.get_nowait()
            i = n.qsize()
        except Exception as e:
            break
        try:
            chrome_options = Options()
            driver = webdriver.Chrome(executable_path = "C:\\Anaconda3\\pkgs\\chromedriver.exe")
            driver.get(url)
            driver.maximize_window()
            driver.minimize_window()
            driver.maximize_window()
            pageSource = driver.page_source

            sp = BeautifulSoup(pageSource, "lxml")
            housedata = sp.select('div.houseList-item-title > a')
            checkdata = sp.select('div.houseList-item-title > span')
            k = ''
            check = []
            for i in range(len(checkdata)):
                for j in checkdata[i].text:
                    if j != ' ':
                        k = k + j
                    else:
                        k = k + j
                        if k == "熱銷建案 ":
                            check.append(i)
                            k = ''
                        elif k == "車位 ":
                            check.append(i)
                            k = ''
                        else:
                            k = ''

            for i in housedata:
                lin = i.get('href')
                link.append(lin)

            driver.close()

            check1 = []
            for i in check:
                check1.append(link[i])

            link1 = []
            for i in link:
                val = 0
                for j in check1:
                    if i != j:
                        continue
                    else:
                        val = val + 1
                if val == 0:
                    link1.append(i)

            for a in range(len(link1)):
                st1 = 'https://sale.591.com.tw'
                st2 = link1[a]
                st = st1 + st2
                link3.append(st)
        except Exception as e:
            continue

if __name__=='__main__':
    time_start=time.time()
    thr = []
    thrnum = 4
    for i in range(thrnum):
        t = threading.Thread(target=getwebsite, args=(urlQueue,))
        thr.append(t)
    for t in thr:
        t.start()
    for t in thr:
        t.join()
    
b = set(link3)
link3 = list(b)

time_end=time.time()
print('time cost',time_end-time_start,'s')
```

```python
link = []
linkrent = []
def getwebsite(n):
    while True:
        try:
            url1 = n.get_nowait()
            i = n.qsize()
        except Exception as e:
            break
        try:
            chrome_options = Options()
            driver = webdriver.Chrome(executable_path = "C:\\Anaconda3\\pkgs\\chromedriver.exe")
            driver.get(url1)
            driver.maximize_window()
            driver.minimize_window()
            driver.maximize_window()
            pageSource = driver.page_source

            sp = BeautifulSoup(pageSource, "lxml")
            housedata = sp.select('li.pull-left.infoContent > h3 > a')
            checkdata = sp.select('li.pull-left.infoContent > p')
            k = ''
            check = []    
            for i in range(len(checkdata)):
                for j in checkdata[i].text:
                    if j != '\n':
                        k = k + j
                    else:
                        if k == '                                                            車位':
                            l = int(i / 3)
                            check.append(l)
                            k = ''
                        else:
                            k = ''

            for i in housedata:
                lin = i.get('href')
                link.append(lin)

            driver.close()

            check1 = []
            for i in check:
                check1.append(link[i])

            link1 = []
            for i in link:
                val = 0
                for j in check1:
                    if i != j:
                        continue
                    else:
                        val = val + 1
                if val == 0:
                    link1.append(i)
                    
            for i in range(len(link1)):
                sp = ''
                for j in link1[i]:
                    if j != ' ':
                        sp = sp + j
                link1[i] = sp 

            for a in range(len(link1)):
                st1 = 'https:'
                st2 = link1[a]
                st = st1 + st2
                linkrent.append(st)
        except Exception as e:
            continue

if __name__=='__main__':
    time_start=time.time()
    thr = []
    thrnum = 4
    for i in range(thrnum):
        t = threading.Thread(target=getwebsite, args=(urlQueue1,))
        thr.append(t)
    for t in thr:
        t.start()
    for t in thr:
        t.join()
    
b = set(linkrent)
linkrent = list(b)

time_end=time.time()
print('time cost',time_end-time_start,'s')
```
